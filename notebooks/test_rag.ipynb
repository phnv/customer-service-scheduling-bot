# RAG Pipeline Manual Testing
This notebook allows you to test the `FAQService` and ChromaDB retrieval directly.

```python
import sys
import os
# Ensure the app module can be found if running from notebooks directory
sys.path.append(os.path.abspath('..'))

from app.services.faq_service import FAQService

# Initialize the FAQ Service (this connects to ChromaDB)
faq_service = FAQService()
print("FAQ Service Initialized successfully!")
```

```python
# Run a test query
query = "What are your business hours?"
print(f"Searching for: {query}\n")

results = faq_service.search_knowledge_base(query, top_k=2)

for idx, res in enumerate(results, 1):
    print(f"--- Result {idx} (Score: {res['score']:.4f}) ---")
    print(f"Title: {res['title']}")
    print(f"Content: {res['content']}\n")
```
