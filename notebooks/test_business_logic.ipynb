## 1. Customer Tools

```python
from app.tools.customer_tools import find_contact_tool, create_contact_tool, find_patient_tool, create_patient_tool

# Find an existing contact (if you seeded the DB)
contact = find_contact_tool.invoke({"phone": "+155500001"})
print("Contact:", contact)
```

```python
# Create a new contact
new_contact_id = create_contact_tool.invoke({
    "full_name": "Jane Smith", 
    "phone": "+155500022", 
    "document": "987654321",
    "email": "jane@example.com"
})
print("New Contact ID:", new_contact_id)
```

## 2. Appointment Tools

```python
from app.tools.appointment_tools import check_availability_tool, reserve_slot_tool, create_appointment_tool

# Check for Cardiology availability
slots = check_availability_tool.invoke({"specialty": "Cardiology"})
print("Available Slots:", slots)
```

```python
# Reserve the first available slot
if slots:
    first_slot_id = slots[0]["slot_id"]
    reservation = reserve_slot_tool.invoke({"slot_id": first_slot_id, "conversation_id": "test-conv-123"})
    print("Reservation:", reservation)
else:
    print("No slots available to reserve.")
```

```python
# Create an appointment from the reservation
if slots:
    appointment = create_appointment_tool.invoke({
        "reservation_id": reservation["reservation_id"], 
        "patient_id": new_patient_id, 
        "contact_id": new_contact_id
    })
    print("Appointment:", appointment)
```

## 3. Qualification Tools

```python
from app.tools.qualification_tools import update_lead_tool, get_lead_tool

# Update lead qualification for a conversation
lead = update_lead_tool.invoke({
    "conversation_id": "test-conv-123", 
    "fields": {
        "summary": "Patient complains of headaches.", 
        "priority": "normal",
        "first_time_patient": True
    }
})
print("Updated Lead:", lead)
```
